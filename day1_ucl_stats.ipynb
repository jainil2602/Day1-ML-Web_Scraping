```python

import requests
from bs4 import BeautifulSoup
import pandas as pd
url = "https://en.wikipedia.org/wiki/European_Cup_and_UEFA_Champions_League_records_and_statistics#Clubs"


headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
all_tables = pd.read_html(response.text)
df = all_tables[0]
 

```

```python
df.rename(columns={"vte Club": "team",
                   "Title(s)": "champion",
                   "Runners-up": "runner_up",
                   "Seasons won": "seasons_won",
                   "Seasons runner-up": "seasons_runnerup",
                  }, inplace=True)
```
